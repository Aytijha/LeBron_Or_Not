```python
import os
scoring_files = os.listdir('/content/scoring_clips')
nonscoring_files = os.listdir('/content/nonscoring_clips')

print(scoring_files)
print(nonscoring_files)
```

```python
import csv
filename = "train.csv"

header = ['file', 'label']
with open(filename, 'w') as csvfile:
  csvwriter = csv.writer(csvfile)
  csvwriter.writerow(header)

  for fil in scoring_files:
    row = [fil, 1]
    csvwriter.writerow(row)
  
  for fil in nonscoring_files:
    row = [fil, 0]
    csvwriter.writerow(row)
```

```python
import pandas as pd
df = pd.read_csv('train.csv')
df
```

```python
import os
test_files = os.listdir('/content/test_clips')

import csv
filename = "test.csv"

header = ['file']
with open(filename, 'w') as csvfile:
  csvwriter = csv.writer(csvfile)
  csvwriter.writerow(header)

  for fil in test_files:
    row = [fil]
    csvwriter.writerow(row)
```

```python
import pandas as pd
df = pd.read_csv('test.csv')
df
```
